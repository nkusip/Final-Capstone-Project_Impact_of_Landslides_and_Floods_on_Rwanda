## **1. Introduction and Background**

Natural disasters represent a growing global challenge, particularly in developing countries where environmental vulnerability is compounded by socio-economic constraints. Among the most frequent and destructive hazards are floods and landslides, which result in significant loss of life, destruction of infrastructure, and disruption of livelihoods. Recent studies have shown that the increasing occurrence of these disasters is closely associated with climate change, environmental degradation, deforestation, and rapid population growth (IPCC, 2022; UNDRR, 2022; World Bank, 2021).

In Rwanda, the risk of natural disasters is amplified by its unique geographical and environmental characteristics (UNDP, 2022; Byiringiro et al., 2024). The country’s mountainous terrain, combined with high rainfall intensity and dense population distribution, makes it particularly susceptible to floods and landslides. These events are often triggered by heavy rains, leading to soil erosion, slope instability, and overflow of water bodies. The Western and Northern Provinces have been consistently identified as the most affected regions, experiencing recurrent disasters that severely impact communities (MIDIMAR, 2015; MINEMA, 2020; World Bank, 2020).

Over the past decade, Rwanda has made significant efforts to strengthen disaster risk management through institutional reforms and early warning systems. However, the increasing frequency and intensity of extreme weather events highlight the need for continuous monitoring and data-driven analysis. Understanding the patterns, impacts, and spatial distribution of floods and landslides is essential for designing effective mitigation and adaptation strategies.

This capstone project therefore seeks to evaluate the impact of floods and landslides in Rwanda over the period of 2021–2025. This study adopts a comprehensive analytical approach, combining descriptive statistics, inferential modeling, and spatial analysis to provide insights into landslide and flood trends and their consequences on human populations and infrastructure.

```r

# Cleaning data 

## Handling missing values

clean_data <- raw_data %>%
         drop_na(,4:21)
summary(clean_data)

        

# Outliers check
# Define outliers using IQR (Interquartile Range)
# calculate IQR (Interquartile Range) for my data

iqr_values <- clean_data %>%
  summarise(across(4:21, ~ IQR(., na.rm = TRUE)))

iqr_values

# Get Q1 and Q3 for each column

iqr_summary <- clean_data %>%
  summarise(across(4:21,
                   list(Q1 = ~quantile(., 0.25, na.rm = TRUE),
                        Q3 = ~quantile(., 0.75, na.rm = TRUE))))

iqr_summary

# Define outlier thresholds

get_iqr_bounds <- function(x, multiplier = 1.5) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR_val <- Q3 - Q1
    
  lower <- Q1 - 1.5 * IQR_val
  upper <- Q3 + 1.5 * IQR_val
  return(c(lower = lower, upper = upper))
}

get_iqr_bounds


# Apply the above function to all columns

iqr_bounds <- lapply(clean_data[4:21], get_iqr_bounds)

iqr_bounds

# Detecting outliers after iqr_bounds

outlier_matrix <- lapply(names(iqr_bounds), function(col) {
  x <- clean_data[[col]]
  bounds <- iqr_bounds[[col]]
  x < bounds["lower"] | x > bounds["upper"]
})

outlier_matrix

names(outlier_matrix) <- names(iqr_bounds)

# Count the nbr of outliers per each column

sapply(outlier_matrix, sum, na.rm = TRUE)

#sapply(outlier_matrix, function(x) sum(x, na.rm = TRUE))


#Outliers check
#Option 1

#outlier_clean_data <- boxplot(clean_data[4:21], main = "Boxplots of Columns 4–21", las = 2, cex.axis = 0.7) 

#outlier_clean_data

#Option 2
#outiers_clean_data <- clean_data %>%
  #select(4:21) %>%             # select columns 4 to 21
  #pivot_longer(everything(), names_to = "Variable", values_to = "Value") %>%
  #ggplot(aes(x = Variable, y = Value)) +
  #geom_boxplot(outlier.colour = "red", outlier.shape = 1) +
  #theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
  #labs(title = "Boxplots of Columns 4–21")

#outiers_clean_data

#Renaming columns


clean_data <-clean_data %>%
    rename (
            `Number of deaths` = Deaths ,
            `Number of injuries` = Injured ,
            `Number of damaged houses` = `Houses Damaged`,
            `Damaged crops (ha)` = `Damages in crops Ha.`,
            `Number of lost cattle` = `Lost Cattle`,
            `Destroyed forest(ha)` = `Forest (Ha)`,
            `Other livestock(number)` = `Other Livestock`,
            `Damaged classrooms(number)` = Classrooms,
            `Damaged health centres (number)` = Healthcenters,
            `Damaged road section(number)` = `Roads Section`,
            `Damaged churches(number)` = Churches ,
            `Damaged bridges (number)` = Bridges,
            `Damaged administrative offices (number)` = `Administrative offices`,
            `Damaged water supply lines(number)` = `Water supply`,
            `Damaged transmission lines(number)` = Markets,
            `Damaged markets(number)` = Markets,
            `Damaged factories (number)` = Factory
    )

clean_data
```

# ANALYSIS , VISUALISATION, AND INTERPRETATION

```r

# ============================================================
# 2. IMPORT AND CLEAN RAW DATA
# ============================================================
raw_data <- read.csv("data/raw/Disaster Impacts 2021_Merged_2025.csv", 
                     check.names = FALSE, 
                     stringsAsFactors = FALSE)

# Remove rows with missing values in key columns (4:21)
clean_data <- raw_data %>% drop_na(4:21)

# Rename columns for easier access (keep original names for backward compatibility)
clean_data <- clean_data %>%
  rename(
    `Number of deaths` = Deaths,
    `Number of injuries` = Injured,
    `Number of damaged houses` = `Houses Damaged`,
    `Damaged crops (ha)` = `Damages in crops Ha.`,
    `Number of lost cattle` = `Lost Cattle`,
    `Destroyed forest(ha)` = `Forest (Ha)`,
    `Other livestock(number)` = `Other Livestock`,
    `Damaged classrooms(number)` = Classrooms,
    `Damaged health centres (number)` = Healthcenters,
    `Damaged road section(number)` = `Roads Section`,
    `Damaged churches(number)` = Churches,
    `Damaged bridges (number)` = Bridges,
    `Damaged administrative offices (number)` = `Administrative offices`,
    `Damaged water supply lines(number)` = `Water supply`,
    `Damaged transmission lines(number)` = Markets,
    `Damaged markets(number)` = Markets,
    `Damaged factories (number)` = Factory
  )

# Note: The column 'Event' remains unchanged

# ============================================================
# 3. CREATE YEARLY AGGREGATES FOR LANDSLIDES AND FLOODS
# ============================================================
summary_all <- clean_data %>%
  filter(Event %in% c("LANDSLIDES", "FLOODS")) %>%
  group_by(Year, Event) %>%
  summarise(
    total_events = sum(`Number of incidents`, na.rm = TRUE),
    deaths = sum(`Number of deaths`, na.rm = TRUE),
    injured = sum(`Number of injuries`, na.rm = TRUE),
    houses = sum(`Number of damaged houses`, na.rm = TRUE),
    crops = sum(`Damaged crops (ha)`, na.rm = TRUE),
    .groups = "drop"
  )

# Split into separate data frames
landslides_df <- summary_all %>% filter(Event == "LANDSLIDES")
floods_df     <- summary_all %>% filter(Event == "FLOODS")

```

```r
# ============================================================
# PREPARE DATA FOR HUMAN IMPACT PLOT
# ============================================================
# Use summary_all (which has columns: Year, Event, deaths, injured)
impact_plot <- summary_all %>%
  rename(disaster = Event) %>%
  select(Year, disaster, deaths, injured) %>%
  pivot_longer(cols = c(deaths, injured),
               names_to = "Impact",
               values_to = "Value") %>%
  group_by(Year, Impact) %>%
  mutate(Percentage = Value / sum(Value, na.rm = TRUE)) %>%
  ungroup()

# ============================================================
# CREATE THE PLOT
# ============================================================
ggplot(impact_plot, aes(x = factor(Year), y = Value, fill = disaster)) +
  
  geom_col(position = position_dodge(width = 0.8)) +
  
  # Value labels above bars
  geom_text(aes(label = Value),
            position = position_dodge(width = 0.8),
            vjust = -0.3,
            size = 3.5) +
  
  # Percentage labels inside bars (only when Value > 0 to avoid warnings)
  geom_text(data = subset(impact_plot, Value > 0),
            aes(label = percent(Percentage, accuracy = 1)),
            position = position_dodge(width = 0.8),
            vjust = 1.5,
            size = 3,
            color = "white") +
  
  facet_wrap(~Impact, scales = "free",
             labeller = labeller(Impact = c("deaths" = "Deaths",
                                            "injured" = "Injuries"))) +
  
  scale_fill_manual(values = c(
    "LANDSLIDES" = "#2C7FB8",
    "FLOODS" = "#F03B20"
  )) +
  
  labs(
    title = "Human Impacts of Landslides and Floods in Rwanda (2021–2025)",
    x = "Year",
    y = "Number of People Affected",
    fill = "Disaster Type"
  ) +
  
  theme_minimal() +
  
  theme(
    plot.title = element_text(face = "bold", size = 14),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 12),
    legend.position = "top"
  )
```

The bar charts illustrate the human impact of landslides and floods in Rwanda between 2021 and 2025, focusing specifically on fatalities and injuries, and revealing strong year-to-year fluctuations with a major crisis peak in 2023.

In terms of fatalities, the early years show relatively low but increasing death counts. In 2021, deaths were minimal, with 3 fatalities caused by floods and 2 by landslides. This increased significantly in 2022, reaching a total of 57 deaths, where floods accounted for 33 deaths (58%) and landslides for 24 deaths (42%). The situation escalated dramatically in 2023, which stands out as the deadliest year in the entire period. Landslides slightly surpassed floods in terms of mortality, causing 87 deaths (54%) compared to 74 deaths (46%) from floods, indicating that both hazards contributed heavily to the humanitarian burden. After this peak, fatalities declined sharply in 2024, with 12 deaths from floods and 14 from landslides. In 2025, however, the pattern shifted again, as floods became the dominant cause of death, accounting for 22 fatalities (96%), while landslides were responsible for only 1 death.

A similar pattern is observed in injury trends, which remained very low in the early years before surging in 2023. In 2021 and 2022, injuries were limited, although landslides were responsible for the majority of cases in 2022, contributing 9 out of the total injuries recorded that year. The year 2023 again represents a severe outlier, with injuries reaching their highest level: floods caused 72 injuries (61%), while landslides caused 47 (39%), showing that floods were the leading driver of non-fatal human harm during that peak disaster year. In 2024, injuries dropped significantly, though landslides accounted for most cases (6 out of 7 total injuries). By 2025, the trend reversed completely, with floods responsible for all recorded injuries (7 cases), while landslides caused none.

Overall, the findings highlight 2023 as a critical humanitarian shock period characterized by extreme fatalities and injuries from both floods and landslides. While landslides were slightly more lethal during the peak year, floods appear to be more consistently associated with injuries and continued human impact in later years, particularly in 2025. The results emphasize the highly volatile nature of disaster impacts, where a single extreme year can dominate long-term human loss trends.

After this peak, both hazards declined significantly in 2024, with flood damage dropping to 125 ha and landslide damage falling to just 12 ha. A similar pattern is observed in housing damage, though with even more pronounced effects in 2023. In 2021, floods damaged 54 houses compared to 11 houses affected by landslides, while in 2022 both hazards caused equal damage of 23 houses each, indicating a relatively balanced impact. The year 2023 stands out dramatically, with floods destroying 2,388 houses and landslides damaging 2,359 houses, representing the most severe residential losses across the entire study period. This sharp increase suggests the occurrence of exceptionally severe disaster events affecting settlements. Following this peak, the situation improved substantially: in 2024, flood damage declined to 26 houses and landslides to 14 houses, while in 2025 the impact became minimal, with only 3 houses affected by floods and 5 by landslides.

Overall, the findings highlight that floods represent a more persistent annual risk to both crops and housing, while landslides exhibit more volatility with extreme spikes, particularly in 2023.

The findings from the analysis of floods and landslides in Rwanda between 2021 and 2025 reveal highly dynamic patterns in disaster occurrence, human impact, spatial distribution, and statistical relationships. Floods peaked in 2022 with 97 incidents, while landslides reached their highest level in 2024 with 111 incidents. Despite these fluctuations, the total distribution over the five‑year period remains nearly balanced, with landslides accounting for 305 incidents (51.9%) and floods for 283 incidents (48.1%). This indicates that Rwanda experiences both hazards at almost equal rates, although their timing and intensity differ considerably across years.

In terms of human impact, floods and landslides together caused 272 casualties, comprising 144 deaths and 128 injuries. Floods were slightly more lethal overall, accounting for a higher total number of deaths across the study period. However, 2023 stands out as a major disaster year, recording 161 deaths in total. During this peak, landslides caused 87 deaths while floods caused 74, showing that both hazards contributed significantly to the humanitarian crisis. Injuries followed a similar pattern, remaining relatively low in most years but increasing sharply in 2023, when both floods and landslides jointly produced the highest burden of non‑fatal impacts. These results demonstrate that although average yearly impacts may appear moderate, extreme events can lead to substantial human losses.

Property and infrastructure damage further highlights the differing nature of these hazards. Approximately 5,129 houses were affected across the study period, with landslides (2,606 houses) slightly exceeding floods (2,523 houses). However, floods had a stronger impact on agriculture, damaging 1,539 hectares of crops compared to 907 hectares affected by landslides. This reflects the tendency of floods to spread across fertile lowlands and river basins. In contrast, landslides were more destructive to infrastructure, particularly roads, damaging 74 road sections compared to only 21 from floods. This distinction shows that floods primarily affect livelihoods and food security, while landslides disrupt transportation and physical connectivity.

The spatial distribution of disasters reveals strong geographic clustering across Rwanda. Landslides are most frequent in Nyamasheke, Rusizi, Ngororero, Rutsiro, and Gakenke, mainly located in the western highland regions. Floods are concentrated in Nyarugenge, Ngororero, Nyamagabe, Gakenke, and Nyabihu. Importantly, districts such as Ngororero, Gakenke, Nyamagabe, and Nyabihu emerge as dual‑hazard hotspots, experiencing high exposure to both floods and landslides. This overlapping risk pattern highlights the need for integrated, multi‑hazard disaster risk management strategies. It is also important to note that this spatial analysis is based on district‑level aggregation rather than full GIS mapping, meaning that it identifies patterns of risk but does not involve geospatial modeling using shapefiles or coordinates.

The statistical modelling results further strengthen the understanding of disaster dynamics. The Poisson regression model shows that disaster occurrence is not significantly influenced by year (β = –0.0326, p = 0.337), indicating no strong temporal trend in incident frequency over the 2021–2025 period. District effects are also not statistically significant in this model (all p‑values > 0.05), suggesting that after controlling for other factors, differences in incident counts across districts are not significant – a finding that contrasts with the raw spatial clustering observed in descriptive statistics. However, the type of event does matter: the coefficient for landslides is positive and significant (β = 0.1724, p = 0.0446), meaning that, holding other variables constant, landslides are associated with a significantly higher number of incidents than floods. This confirms that while floods and landslides occur with similar total frequency, landslides are more frequent per event category.

The hypothesis test on whether landslides are more deadly than floods shows that the null hypothesis cannot be rejected. Although landslides were more deadly in 2023, floods caused more total deaths over the five‑year period (144 compared to 128). The linear regression model (model1) further supports this: the coefficient for EventLANDSLIDES is not statistically significant (β = –0.135, p = 0.629), indicating that, after accounting for the number of incidents, houses damaged, and crop damage, there is no meaningful difference in deaths between the two hazard types.

Finally, the regression models identifying drivers of disaster impacts show that structural damage – particularly the number of damaged houses – and the number of incidents are strong predictors of mortality. The full model (model1) explains 80.9% of the variation in deaths (adjusted R² = 0.8043), confirming a very strong relationship between disaster intensity and human outcomes. Separate models for floods and landslides reveal even higher explanatory power (adjusted R² = 0.8205 for floods, 0.9351 for landslides). Notably, each additional damaged house contributes roughly twice as much to landslide‑related deaths (β = 0.034) as to flood‑related deaths (β = 0.017), while each additional incident has a larger marginal effect on flood deaths (β = 0.379) than on landslide deaths (β = 0.139). This nuanced finding underscores that flood mortality is driven more by event frequency, whereas landslide mortality is driven more by the severity of structural damage. Overall, the results demonstrate that disaster impacts in Rwanda are shaped by a combination of temporal variability, spatial concentration, and hazard‑specific characteristics, highlighting the need for targeted, location‑specific, and multi‑hazard disaster preparedness strategies.


## *Conclusion and Recommendations*

### Conclusion

This study analyzed the trends, impacts, spatial distribution, and statistical drivers of floods and landslides in Rwanda between 2021 and 2025. The findings reveal that both hazards occur with nearly equal frequency over time, with landslides accounting for 51.9% of incidents and floods for 48.1%. However, their impacts differ significantly depending on the type of exposure and year of occurrence. Floods were found to be slightly more lethal overall in terms of total deaths, while landslides caused more concentrated damage during extreme events such as in 2023.

The year 2023 emerged as a critical turning point, recording the highest number of deaths (161) and injuries, confirming it as a major disaster year for Rwanda. Property and infrastructure damage was also substantial, with over 5,000 houses affected and significant losses in crops and road networks. Floods were more destructive to agricultural land, while landslides caused more severe damage to infrastructure such as roads and bridges.

Spatially, disaster risk is not evenly distributed across the country. Western and mountainous districts such as Ngororero, Rusizi, Nyamasheke, Gakenke, and Nyabihu consistently experience higher exposure to both hazards, indicating strong geographic clustering of risk. These areas represent multi‑hazard hotspots that require prioritized intervention.

Statistically, the models confirm that disaster impacts are strongly influenced by event frequency and structural damage. The number of incidents and the number of damaged houses together explain over 80% of the variation in deaths. While landslides are significantly more frequent than floods in the Poisson model, they are not significantly deadlier once other factors are controlled for. The separate models for each hazard reveal that flood deaths are more sensitive to incident frequency, whereas landslide deaths are more sensitive to housing damage. Overall, the study confirms that disaster risk in Rwanda is complex, spatially concentrated, and highly sensitive to extreme weather conditions and infrastructure vulnerability.

## Recommendations

Based on the findings of this study, several key recommendations are proposed to reduce disaster risk and improve resilience in Rwanda.

First, there is a need to strengthen early warning systems and climate monitoring, particularly in high‑risk districts. Since floods and landslides are strongly linked to heavy rainfall and extreme weather events, improving real‑time weather forecasting and community alert systems would help reduce fatalities and injuries. The finding that flood deaths are driven more by incident frequency underscores the value of timely evacuation warnings.

Second, the government and relevant agencies should prioritize targeted infrastructure reinforcement in high‑risk districts such as Ngororero, Rusizi, Nyamasheke, Gakenke, and Nyabihu. Given that landslide deaths are highly sensitive to housing damage, investments in slope stabilization, retaining walls, and relocation of settlements from steep slopes will yield large reductions in mortality. For flood‑prone areas, improved drainage and flood‑proofing of buildings are essential.

Third, community‑based disaster preparedness programs should be expanded. Local populations in hotspot districts should be trained on evacuation procedures, emergency response, and risk awareness to reduce vulnerability during disaster events. The high explanatory power of the regression models suggests that simple, observable indicators (e.g., number of houses damaged) can be used to trigger community‑level actions.

Fourth, there is a need to invest in climate‑resilient agriculture, particularly in flood‑prone areas. Since floods significantly affect crop production, introducing flood‑resistant farming techniques and improved drainage in agricultural zones can reduce food insecurity. The substantial crop damage caused by floods (1,539 ha) relative to landslides (907 ha) justifies this focus.

Fifth, the study recommends improving data collection and integration systems, including the incorporation of geospatial (GIS) data and meteorological datasets. While the current district‑level aggregation revealed clear spatial patterns, the Poisson model did not find significant district effects after controlling for other variables – possibly due to aggregation or omitted confounders. Finer‑scale, georeferenced data would enable more precise risk mapping and better prediction of disaster‑prone areas.

Finally, future research should incorporate rainfall and climate variability data to better understand the direct relationship between weather patterns and disaster occurrence, as this was not fully captured in the current dataset. Additionally, the observed volatility in annual incidents (e.g., flood peak in 2022, landslide peak in 2024) calls for investigation into climatic drivers such as the Indian Ocean Dipole or El Niño–Southern Oscillation.

Overall, implementing these recommendations would significantly strengthen Rwanda’s disaster risk reduction framework and reduce both human and economic losses associated with floods and landslides.
